#Tune plots
In order to print the plots, input the alg name and parameters. Then the plots for that family will be shown. 

```python
from helper import get_formatted_name
params = {
    'env_name':'FetchPush-v1',
    'alg':'naf',
    'exp_name':'AL47',
    'obs_noise_std':0.0,
    'action_noise_std':0.0,
    'goal_radius':0.05
}
run_path = '/home/gridsan/alagrassa/git/baselines/run_results/'
name = get_formatted_name(params)
print(name)
import sys
sys.path.append("/home/gridsan/alagrassa/git/rl_plot/rl_plot")
import make_plots as mp
import numpy as np
import matplotlib.pyplot as plt
reload(mp)
metrics = {'ppo2':['approxkl', 'clipfrac', 'exp_variance', 'policy_entropy', 'policy_loss', 'value_loss'],
           'ddpg':['action_mean', 'loss_actor', 'loss_critic', 'param_noise_distance'],
           'cma':['guess_accuracy', 'mueff', 'sigma', 'weights']}
colors = ['r', (1, 166/255., 0, 1), 'y', 'g', 'b', (149/255.,0,1,1)]
def plot_goal_radiuses(params, goal_radiuses=None, cutoff=None, eps_per_iter=1, data_cutoff=None):
    if goal_radiuses is None:
        goal_radiuses = [0.05, 0.08, 0.01, 0.02, 0.03, 0.1]
    else:
        print(goal_radiuses)
        
    goal_radiuses.sort()
    actual_name = params["exp_name"]
    for goal_radius, color in zip(goal_radiuses, colors):
        params["goal_radius"] = goal_radius
        name = get_formatted_name(params)
        mp.plot_family(run_path+name+"test_success_rates", eps_per_iter=eps_per_iter, data_cutoff=data_cutoff, smoothing=2, display=False, 
                       label="goal radius:"+str(goal_radius), color=color, cutoff=cutoff, ylab="mean reward")
        plt.legend()
        plt.show()

    for metric in metrics[params['alg']]:
        mp.plot_family(run_path+name+metric, eps_per_iter=eps_per_iter, data_cutoff=data_cutoff, smoothing=2, display=False, 
                       ylab="metric:"+str(metric), color='b', cutoff=cutoff)
        plt.legend()
        plt.show()
    
    
        
    #mp.plot_family(run_path+name+"test_success_rates", cutoff=cutoff,smoothing=2, data_cutoff=data_cutoff, eps_per_iter=eps_per_iter, display=False, 
    #                   label="action noise radius:"+str(goal_radius), color=color)
    
def plot_actuation_noise(params, eps_per_iter=1, cutoff=None, goal_radiuses=None, data_cutoff=None):
    if goal_radiuses is None:
        goal_radiuses = [0.0, 0.01, 0.1, 0.2, 0.05, 0.5]
    goal_radiuses.sort()
    actual_name = params["exp_name"]
    for goal_radius, color in zip(goal_radiuses, colors):
        params["action_noise_std"] = goal_radius
        name = get_formatted_name(params)
        mp.plot_family(run_path+name+"test_success_rates", cutoff=cutoff,smoothing=1, data_cutoff=data_cutoff, eps_per_iter=eps_per_iter, display=False, 
                       label="action noise radius:"+str(goal_radius), color=color)
    plt.legend()
    plt.show()

def plot_observation_noise(params, eps_per_iter=1, cutoff=None, goal_radiuses=None, data_cutoff=None):
    if goal_radiuses is None:
        print("goal radiuses was none")
        goal_radiuses = [0.0, 0.01, 0.05, 0.1, 0.2, 0.5]
    goal_radiuses.sort()
    actual_name = params["exp_name"]
    for goal_radius, color in zip(goal_radiuses, colors):
        params["obs_noise_std"] = goal_radius
        name = get_formatted_name(params)
        mp.plot_family(run_path+name+"test_success_rates", cutoff=cutoff,smoothing=2, data_cutoff=data_cutoff,  display=False, 
                       label="obs noise:"+str(goal_radius), eps_per_iter=eps_per_iter, color=color)
    plt.legend()
    plt.show()
```
