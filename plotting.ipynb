#Tune plots
In order to print the plots, input the alg name and parameters. Then the plots for that family will be shown. 

```python
from helper import get_formatted_name
params = {
    'env_name':'FetchPush-v1',
    'alg':'naf',
    'exp_name':'AL47',
    'obs_noise_std':0.0,
    'action_noise_std':0.0,
    'goal_radius':0.05,
    'rew_noise_std':5
}
run_path = '/home/gridsan/alagrassa/git/baselines/run_results/'
name = get_formatted_name(params)
print(name)
import sys
sys.path.append("/home/gridsan/alagrassa/git/rl_plot/rl_plot")
import make_plots as mp
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
#reload(mp)
metrics = {'ppo2':['approxkl', 'clipfrac', 'exp_variance', 'policy_entropy', 'policy_loss', 'value_loss', 'actualloss'],
           'ddpg':['action_mean', 'loss_actor', 'loss_critic', 'param_noise_distance'],
           'cma':['mueff', 'sigma', 'weights']}
colors = ['r', (1, 166/255., 0, 1), 'y', 'g', 'b', (149/255.,0,1,1)]

def plot_changing_factor(params, change_factor, show_int = False, special_cases = {}, goal_radiuses=None,
                         cutoff=None, eps_per_iter=1, data_cutoff=None, smoothing=5):
    input_goal_radiuses = goal_radiuses
    if change_factor == 'goal_radius':
        goal_radiuses = [0.05, 0.08, 0.01, 0.02, 0.03, 0.1]
        nice_label = "goal radius"
    elif change_factor == "action_noise_std":
        goal_radiuses = [0.0, 0.1, 0.5, 1.0, 2.0]
        nice_label = "action noise"
    elif change_factor == "obs_noise_std":
        goal_radiuses = goal_radiuses = [0.0, 0.01, 0.05, 0.1, 0.2, 0.5]
        nice_label = "observation noise"
    elif change_factor == "rew_noise_std":
        goal_radiuses = [0.,0.25, 0.5, 0.75, 1.]
        nice_label = "reward noise"
    else:
        raise ValueError("Not a valid change factor")
    if input_goal_radiuses is not None:
        goal_radiuses = input_goal_radiuses
    goal_radiuses.sort()
    actual_name = params["exp_name"]
    for goal_radius, color in zip(goal_radiuses, colors):
        params[change_factor] = goal_radius
        if goal_radius in special_cases:
            params['exp_name']=special_cases[goal_radius]
            
        name = get_formatted_name(params)
        mp.plot_family(run_path+name+"test_success_rates", 
                       eps_per_iter=eps_per_iter, data_cutoff=data_cutoff, 
                       smoothing=smoothing, display=False, label=nice_label+":"+str(goal_radius), 
                       color=color, cutoff=cutoff, ylab="success rate")
        params['exp_name'] = actual_name
    #plt.ylim([0,1])
    plt.legend()
    plt.show()
    if show_int:
        for metric in metrics[params['alg']]:
            for goal_radius, color in zip(goal_radiuses, colors):
                params[change_factor] = goal_radius
                if goal_radius in special_cases:
                    params['exp_name']=special_cases[goal_radius]
                name = get_formatted_name(params)
                mp.plot_family(run_path+name+metric,
                               label = nice_label+":"+str(goal_radius), 
                               eps_per_iter=eps_per_iter, data_cutoff=data_cutoff,
                               smoothing=smoothing, display=False, 
                            ylab="metric:"+str(metric), color=color, cutoff=cutoff)
                params['exp_name'] = actual_name
            plt.legend()
            plt.show()
```
